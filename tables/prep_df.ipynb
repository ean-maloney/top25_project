```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("2021AP.xlsx")
```

```python
df.drop(index = len(df['Team']) - 1, inplace = True)
```

```python
df
```

```python
week_num = max(pd.to_numeric(df['Week']))
week_num
```

```python
for team in df['Team']:
    for week in range(1, week_num + 1):
        if df.loc[(df["Team"] == team) & (df["Week"] == week)].size == 0:
            df.loc[len(df.index)] = [team, week, 26, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
```

```python
df
```

```python
df.to_csv(f'thru_week_{week_num}.csv', header = True, index = False)
```
