```python
import pandas as pd
import experimental_models as em
```

```python
mlr_params = {}
mlr_params['feature_drop'] = ['Movement', 'Team', 'Points Scored']
rf_params = {}
rf_params['feature_drop'] = ['Movement', 'Team', 'Points Scored']
ab_params = {}
ab_params['feature_drop'] = ['Movement', 'Team', 'Points Scored']

next_week = 15
```

```python
df = em.data_prep(next_week)
df
```

```python
lin_df = pd.DataFrame()
lin_r2 = []
lin_abs = []
lin_week = []

for week in range(5, next_week):
    df = em.data_prep(week)
    
    lin_week.append(week)
    
    r2, _abs = em.run_mlr(mlr_params, df, week)
    lin_r2.append(r2)
    lin_abs.append(_abs)
    
    print(r2, _abs)

lin_df['Week'] = lin_week
lin_df['R2'] = lin_r2
lin_df['Abs'] = lin_abs
```

```python
lin_df
```

```python
rf_df = pd.DataFrame()
rf_r2 = []
rf_abs = []
rf_week = []

for week in range(5, next_week):
    df = em.data_prep(week)
    
    rf_week.append(week)
    
    r2, _abs = em.run_rf(rf_params, df, week)
    rf_r2.append(r2)
    rf_abs.append(_abs)
    
    print(r2, _abs)
    
rf_df['Week'] = rf_week
rf_df['R2'] = rf_r2
rf_df['Abs'] = rf_abs
```
